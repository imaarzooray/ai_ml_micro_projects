```python
import numpy as np
```

```python
# inputs (dendrites)
x = np.array([1.0,2.0])

# target value
y = 1

# weights
w = np.array([0.2,-0.5])

# bias
b = 0.1

# learning rate (eata)
eta = 0.1
```

```python
for epoch in range(1,4):
        print(f"Epoch - {epoch} ")
        
        # Summation [ Net Input ]
        # ? Z = E(Wi*Xi) + bias 
        z  = np.dot(x,w) + b

        # ? Activation Function
        y_pred = 0 if z < 0 else 1

        # ? Error Calculation
        error = y - y_pred


        print(f"Summation of Inputs: {z}")
        print(f"Predicted Output: {y_pred}")
        print(f"Error: {error}")



        # in case of error, we have to update the weights and bias
        #? weight_new = weight_old + change in weights
        #? change_in_weights = eta(error)x

        w = w + eta*error*x

        #? bias_new = old_bias + change in bias
        #? change_in_bias = eta(error)

        b = b + eta*error
        print("\n")
```
